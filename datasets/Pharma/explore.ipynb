```python
import numpy as np
import pandas as pd
import random
import pickle 
```

```python
#Reading all the data
data = pd.read_csv("raman_spectra_api_compounds.csv", header=0)
```

```python
mapping = {} #material_name -> [spectra]
new_x_axis = np.linspace(0,3275,1024)
old_x_axis = np.array([i for i in range(3276)])

for i in range(len(data)):
    z = np.interp(new_x_axis,old_x_axis,np.array(data.iloc[i][:-1],dtype="float"))[np.newaxis, ...] #of shape (1,1024)
    if data["label"].iloc[i] not in mapping:
        mapping[data["label"].iloc[i]] = [z]
    
    else:
        mapping[data["label"].iloc[i]].append(z)
    
```

```python
min([len(v) for _,v in mapping.items()])
```

```python
for k,v in mapping.items():
    print(k,len(v))
```

```python
#The test dataset has 22 samples each for every material in the dataset
#The train dataset has the remaining samples
#This is approximately a 80:20 train-test split

train_mapping = {}
test_mapping = {}
random.seed(42)

for k,v in mapping.items():
    indices = [i for i in range(len(v))]
    random.shuffle(indices)

    test_mapping[k] = [mapping[k][i] for i in indices[:22]]
    train_mapping[k] = [mapping[k][i] for i in indices[22:]]
```

```python
train_spectra = []
train_labels = []

for k,v in train_mapping.items():
    train_spectra.extend(v)
    train_labels.extend([k for _ in range(len(v))])

test_spectra = []
test_labels = []

for k,v in test_mapping.items():
    test_spectra.extend(v)
    test_labels.extend([k for _ in range(len(v))])
```

```python
print(f"The size of the train dataset is {len(train_spectra)}")
print(f"The size of the test dataset is {len(test_spectra)}")
```

```python
with open("Pharma_train.pkl","wb") as file:
    pickle.dump([train_labels,train_spectra],file) 

with open("Pharma_test.pkl","wb") as file:
    pickle.dump([test_labels,test_spectra],file) 
```
